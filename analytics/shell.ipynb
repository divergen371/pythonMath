```python
def fib(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
fib(30)
```

```python
from ipywidgets import interact

f = lambda x: x**3 - x + 1
g = lambda x: 3 * x**2 - 1
f2 = lambda a, x: g(a) * (x - a) + f(a)


def mydraw(contact_point_of_x, color):
    x = np.linspace(-2.0, 2.0)
    y1 = f(x)
    y2 = f2(contact_point_of_x, x)
    plt.figure(figsize=(20,10))
    plt.plot(x, y1)
    plt.plot(x, y2, color=color[0])


interact(mydraw, contact_point_of_x=(-2.0, 2.0), color=["red", "green"])
```
